```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv("carprices.csv")
df.head()
```

### Pandas Dummies Method

```python
dummies = pd.get_dummies(df["Car Model"])
dummies.head()
```

```python
df_dummies = pd.concat([df, dummies], axis = 1)
df_dummies.head()
```

```python
final = df_dummies.drop(["Car Model", "Mercedez Benz C class"], axis = 1)
final.head()
```

```python
X = final.drop("Sell Price($)", axis = 1)
X.head()
```

```python
y = final["Sell Price($)"]
y.head()
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
model.predict([[40000, 1, 0, 1]])
```

```python
model.score(X, y)
```

### Scikit-learn One Hot Encoding Method

```python
le = LabelEncoder()
dfle = df
dfle["Car Model"] = le.fit_transform(dfle["Car Model"])
dfle
```

```python
X = dfle[["Car Model", "Mileage", "Age(yrs)"]].values
X
```

```python
y = dfle["Sell Price($)"]
y
```

```python
ct = ColumnTransformer([("ohe", OneHotEncoder(), [0])], remainder='passthrough')
X = ct.fit_transform(X)
X
```

```python
X = X[:, 1:]
X
```

```python
model.fit(X, y)
```

```python
model.predict([[1, 0, 40000, 1]])
```

```python
model.score(X, y)
```
